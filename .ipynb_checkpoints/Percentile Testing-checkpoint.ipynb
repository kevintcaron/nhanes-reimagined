```python
# Blood Total Mercury
df = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/PBCD_H.XPT')
demo = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/DEMO_H.XPT')
df = pd.merge(df, demo, on='SEQN', how='inner')

var = 'LBXTHG'
```

```python
# Create groupings

# Define age groups
conditions = [
    df['RIDAGEYR'].apply(lambda x: 1 <= x < 6),
    df['RIDAGEYR'].apply(lambda x: 6 <= x < 12),
    df['RIDAGEYR'].apply(lambda x: 12 <= x < 20)]

choices = [1,2,3]

df['age_grp'] = np.select(conditions, choices, default=4)

# df['age_grp'] = pd.cut(df['RIDAGEYR'], bins=[5, 11, 19, float('inf')], labels=[1, 2, 3], right=True)
df['age_group'] = (df['RIDAGEYR'] >= 6).astype(int)

# Define gender
df['gender'] = df['RIAGENDR'].map({1: 1, 2: 2})

# Define race
df['race'] = df['RIDRETH3'].map({
    1: 1,  # 'MA'
    # 2: 2,  # 'OH' - commented out as per original code comment
    3: 2,  # 'NHW'
    4: 3,  # 'NHB'
    6: 4,  # 'NHA'
    7: 5   # Non-Hispanic Multi-racial
})

# Define racial
df['racial'] = df['RIDRETH3'].apply(lambda x: 1 if x in [1, 2] else 2)
```

```python
biomarker_stratified_mean = TaylorEstimator("mean")

biomarker_stratified_mean.estimate(
    y=np.log(df[var]),
    # samp_weight=df["WTMEC2YR"],
    samp_weight=df["WTSH2YR"], # Mercury
    stratum=df["SDMVSTRA"],
    psu=df["SDMVPSU"],
    remove_nan=True)
```

```python
result = biomarker_stratified_mean.to_dataframe()
result['_estimate'] = np.e**result['_estimate']
result['_lci'] = np.e**result['_lci']
result['_uci'] = np.e**result['_uci']
```

```python
pd.set_option('display.max_rows', None)
result['cumulative'] = result['_estimate'].cumsum()
# result['cum_lci'] = result['_lci'].cumsum()
# result['cum_uci'] = result['_uci'].cumsum()


# quantiles = [0, 0.01, 0.05, 0.10, 0.25, 0.50, 0.75, 0.90, 0.95, 0.99, 1]
quantiles = [1, 0.99, 0.95, 0.90, 0.75, 0.50, 0.25, 0.10, 0.05, 0.01, 0]

for quantile in quantiles:
    for i in range(len(result)):
        if result['cumulative'][i] >= quantile:
            print(f'{quantile}% {result['_level'][i]}')
            break
```

```python
def find_percentile(df, percentile):
    target = percentile / 100.0
    return df[df['cumulative'] >= target]['_level'].iloc[0]

# Find the 50th percentile (median)
median_level = find_percentile(result, 50)
print(f'The magnitude at the 50th percentile is: {median_level}')
```

```python
# Compute confidence intervals using Korn and Graubard method
alpha = 0.05
z_score = np.abs(np.percentile(np.random.standard_normal(1000000), [alpha/2, 1 - alpha/2]))

# Get the standard error for the median level
stderr = result[result['_level'] == median_level]['_stderror'].iloc[0]

# Compute the confidence interval for the median
lower_ci = median_level - z_score[1] * stderr
upper_ci = median_level + z_score[1] * stderr

print(f'Confidence interval for the 50th percentile: ({lower_ci}, {upper_ci})')
```

```python
import scipy.stats as stats
# Calculate the confidence intervals for the proportion
alpha = 0.05
z = stats.norm.ppf(1 - alpha / 2)
# z = np.abs(np.percentile(np.random.standard_normal(1000000), [alpha/2, 1 - alpha/2]))


# Calculate the cumulative sum of the variance
result['cumulative_var'] = (result['_stderror'] ** 2).cumsum()

# Find the standard error for the 50th percentile
std_error_at_percentile = np.sqrt(result[result['cumulative'] >= 0.5]['cumulative_var'].iloc[0])

# Compute the proportion at 50th percentile
proportion = result[result['cumulative'] >= 0.5]['cumulative'].iloc[0]

# Calculate the CI for the proportion
lower_ci_proportion = proportion - z * std_error_at_percentile
upper_ci_proportion = proportion + z * std_error_at_percentile

# Transform back to magnitude (level) using interpolation
lower_ci_level = np.interp(lower_ci_proportion, result['cumulative'], result['_level'])
upper_ci_level = np.interp(upper_ci_proportion, result['cumulative'], result['_level'])

print(f'Confidence interval for the 50th percentile: ({lower_ci_level}, {upper_ci_level})')
```

```python
def recode_df_domains(df):
    # Map number to Race
    race_mapping = {
        1.0: 'Mexican American',
        2.0: 'Other Hispanic',
        3.0: 'Non-Hispanic White',
        4.0: 'Non-Hispanic Black',
        5.0: 'Other Race - Including Multi-Racial'}
    df['RIDRETH1'] = df['RIDRETH1'].replace(race_mapping)

    # Map number to Gender
    gender_mapping = {
        1.0: 'Male',
        2.0: 'Female'}
    df['RIAGENDR'] = df['RIAGENDR'].replace(gender_mapping)

    # Create Youth Age Groups
    conditions = [
        df['RIDAGEYR'].apply(lambda x: 3 <= x < 6),
        df['RIDAGEYR'].apply(lambda x: 6 <= x < 12),
        df['RIDAGEYR'].apply(lambda x: 12 <= x < 20)]

    choices = [
        'a. Age 03-05 years',
        'b. Age 06-11 years',
        'c. Age 12-19 years']

    df['Youth Age Group'] = np.select(conditions, choices, default='d. Age 20+ years')

    return df
```

```python
df = recode_df_domains(df)
```

```python
# Testing Prportions with Domains:

cot_mean_prop = TaylorEstimator("proportion")

cot_mean_prop.estimate(
    y=df[var],
    # samp_weight=df["WTMEC2YR"],
    samp_weight=df["WTSH2YR"], # Mercury
    stratum=df["SDMVSTRA"],
    psu=df["SDMVPSU"], 
    domain=df['Youth Age Group'],
    remove_nan=True)
```
